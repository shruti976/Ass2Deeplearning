```python
import numpy as np
import random

def sigmoid(x):
    y = 1/(1 + np.exp(-x))
    return y


t=[]
for i in range(0,100):
    t.insert(i, 0)


for i in range(0,100):
    if i< 50:
        t[i] = 0
    else:
        t[i] = 1
#print("t:", t)

data1 = np.random.uniform(0,3,size=(50,2))
data2 = np.random.uniform(6,9,size=(50,2))
#print("data1:", data1)
#print("data2:", data2)

data = np.append(data1,data2,axis=0)
#print("input data:", data)

w1 = [[random.uniform(-2.0,2) for i in range(2)] for j in range(4)] 
b1 = random.uniform(-2.0,2) 
b2 = random.uniform(-2.0,2)
b3 = random.uniform(-2.0,2)
b4 = random.uniform(-2.0,2) 
    
w2 = [[random.uniform(-2.0,2) for j in range(4)] for k in range(1)]
b5 = random.uniform(-2.0,2) 


lr = 0.01 
epochs = 10000  
error_thresh = 0.01   
count = 1


while count < epochs:  
    output = []  
    err = 0
    dw1 = [[0 for i in range(2)] for j in range(4)] 
    
    for n in range(0,100):
        v1 = (data[n][0] * w1[0][0]) + (data[n][1] * w1[0][1]) + b1
        v2 = (data[n][0] * w1[1][0]) + (data[n][1] * w1[1][1]) + b2
        v3 = (data[n][0] * w1[2][0]) + (data[n][1] * w1[2][1]) + b3
        v4 = (data[n][0] * w1[3][0]) + (data[n][1] * w1[3][1]) + b4

        z = []
        z.append(sigmoid(v1))
        z.append(sigmoid(v2))
        z.append(sigmoid(v3))
        z.append(sigmoid(v4))

        u = (z[0] * w2[0][0]) + (z[1] * w2[0][1]) + (z[2] * w2[0][2]) + (z[3] * w2[0][3]) + b5

        y = sigmoid(u)

        output.insert(n,y)
        
        dw200 = lr * z[0] * (t[n]-y) * y*(1-y)
        dw210 = lr * z[1] * (t[n]-y) * y*(1-y)
        dw220 = lr * z[2] * (t[n]-y) * y*(1-y)
        dw230 = lr * z[3] * (t[n]-y) * y*(1-y)

        db5 = lr * (t[n]-y) * y*(1-y)
        
        w2[0][0] = w2[0][0] + dw200
        w2[0][1] = w2[0][1] + dw210
        w2[0][2] = w2[0][2] + dw220
        w2[0][3] = w2[0][3] + dw230
        b5 = b5 + db5
        
        
        summation = 0
        summation = (w2[0][0]*y*(1-y)*(t[n]-y)) + (w2[0][1]*y*(1-y)*(t[n]-y)) + (w2[0][2]*y*(1-y)*(t[n]-y)) + (w2[0][3]*y*(1-y)*(t[n]-y))

        for i in range(0,2):  
            for j in range(0,4):  
                dw1[j][i] = lr * data[n][i] * z[j]*(1-z[j]) * summation
        
        db1 = lr * z[0]*(1-z[0]) * summation
        db2 = lr * z[1]*(1-z[1]) * summation
        db3 = lr * z[2]*(1-z[2]) * summation
        db4 = lr * z[3]*(1-z[3]) * summation
            

        for i in range(0,2):  
            for j in range(0,4):  
                w1[j][i] = w1[j][i] + dw1[j][i]
        b1= b1 + db1
        b2= b2 + db2
        b3= b3 + db3
        b4= b4 + db4
    
    for n in range(0,len(output)):
        err = err + (t[n] - output[n])**2
    if err < error_thresh:
        break
    else:
        print("Iteration:", count)
        print("\tError:", err)
        count = count + 1 

print("****************************************************************")
print("\nTotal training iteration: ", count-1)
print("\nInput data: ", data)
print ("\nTarget output: ", t)
print ("\nPredicted Output y: ", output)
print("\nWeights of the first layer:")
for i in range(0,2):  
    for j in range (0,4):  
        print("W%d%d: %f" % (i,j, w1[j][i]))
        
print("Biases of the first layer:")       
print("b1: ",b1)
print("b2: ",b2)
print("b3: ",b3)
print("b4: ",b4)

print("\nWeights of the second layer:")
for j in range (0,4):  
    for k in range (0,1):  
        print("W'%d%d: %f" % (j,k,w2[0][j]))
        
print("Bias of the second layer:")  
print("b5: ", b5)
```
